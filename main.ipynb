```python
final_submit = pd.DataFrame()
final_submit['userid'] = data9['userid']
```

```python
rr = data20
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer2 = CountVectorizer(max_features=7000, min_df=100, max_df=.30)
X = vectorizer2.fit_transform(data20.url).toarray()
```

```python
print(X.shape)
type(X)
```

```python
#Y = rr.gender
rr.iloc[0][1]
```

```python
YY = []
for i in range(0,len(rr)):
  if(rr.iloc[i][1]=='M'):
    YY.append(1)
  else:
    YY.append(0)
```

```python
Y = np.asarray(YY)
type(Y[0])
```

```python
vvv = vv*len(data9)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Dropout

layers = [
    Dense(1000, input_shape=(7000,), activation='relu'),
    Dropout(.5),
    Dense(300, activation='relu'),
    Dropout(.5),
    Dense(120, activation='relu'),
    Dropout(.5),
    Dense(80, activation='relu' ,kernel_regularizer = regularizers.l2(.01)),
    Dense(20, activation='relu' ),
    Dense(2, activation='softmax')
]
model = Sequential(layers)
```

```python
model.compile(
    Adam(lr=.0003), 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
```

```python
scaled_train_samples = X_train
train_labels = Y_train
```

```python
model.fit(
    scaled_train_samples, 
    train_labels, 
    batch_size=128, 
    epochs=500,
    validation_data=(X_test, Y_test), 
    shuffle=True, 
    verbose=2
)
```
